# menu 크롤링

```python
menu_list=[]
infos=[]

driver = webdriver.Chrome()
for page in range(1,10):
    url = f"https://composecoffee.com/menu?page={page}"
    driver.get(url)
    response=requests.get(url)
    time.sleep(1)
    button = driver.find_elements(By.TAG_NAME,'button')[1].click()
    soup=BeautifulSoup(response.text, 'html.parser')
    time.sleep(1)
    menus = [menu.text for menu in driver.find_elements(By.CLASS_NAME,'undertitle')]
    menu_list.extend(menus)
    print(menus)
    info2 = soup.find_all('div',attrs={'class':'col-lg-3 col-md-4 col-sm-6 itemBox'})
    infos.extend(info2)
    print(info2)
driver.close()
```

```python
info=[]
for i in range(len(infos)):
    menu_info=[]
    inf=infos[i].find_all('li',attrs={'class':'extra'})
    for j in range(len(inf)):   
        menu_info.append(inf[j].text)
    info.append(menu_info)
```

# DB 등록

```python
conn = sqlite3.connect("Comfile_Coffee_DB.db")
cursor = conn.cursor()

# menus 테이블 생성
cursor.execute('''
    CREATE TABLE IF NOT EXISTS menus (
        menu_id INTEGER PRIMARY KEY AUTOINCREMENT,
        menu_name TEXT NOT NULL,
        category TEXT NOT NULL,
        info TEXT
    )
''')
conn.commit()
```

```python
for key, value in menu_dict.items():
    menu_name = key
    category = value.pop()
    info =" ".join(value)
    cursor.execute("INSERT OR REPLACE INTO menus (menu_name, category, info) VALUES (?, ?, ?)", 
                   (menu_name , category , info))
conn.commit()
cursor.close()
```

```python
cursor = conn.cursor()

# users 테이블 생성
cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        user_id INTEGER PRIMARY KEY AUTOINCREMENT,
        face_encoding BLOB 
        name TEXT NOT NULL,       
        phone TEXT ,
        created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
    )
''')

# orders 테이블 생성
cursor.execute('''
    CREATE TABLE IF NOT EXISTS orders (
        order_id INTEGER PRIMARY KEY AUTOINCREMENT,
        user_id INTEGER,
        total_price REAL,
        ordered_menu TEXT,
        status TEXT CHECK(status IN ('complete','cancel','refund','pending')),
        ordered_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
        FOREIGN KEY (user_id) REFERENCES users(user_id) ON DELETE SET NULL
    )
''')


conn.commit()
cursor.close()

```
